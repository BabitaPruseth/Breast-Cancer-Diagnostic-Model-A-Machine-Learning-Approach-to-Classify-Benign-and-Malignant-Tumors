#### 4. Data Preprocessing

```python
import matplotlib.pyplot as plt

# Convert the 'diagnosis' column into binary values (M -> 1 for malignant, B -> 0 for benign)
df['diagnosis'] = df['diagnosis'].apply(lambda val: 1 if val == 'M' else 0)

# Plot the count of each category in the 'diagnosis' column
plt.figure(figsize=(8, 5))  # Adjust the width and height of the figure
df['diagnosis'].value_counts().plot(kind='bar', color=['teal', 'salmon'], edgecolor='black', width=0.4)
plt.title('Diagnosis Distribution (M=1, B=0)', fontsize=14)
plt.xlabel('Diagnosis', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.show()
```

```python
# Remove ID column for analysis 
df.drop('id', axis=1, inplace=True)
```

##### 6. Feature Selection

```python
# Calculate correlation matrix for feature selection
corr_matrix = df.corr().abs()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)

# Identify highly correlated features
to_drop = [x for x in tri_df.columns if any(tri_df[x]>0.92)]
df = df.drop(to_drop, axis=1)

print(df.shape[1])
```

```python
df.head()
```

#### Here, reduced the feature set from 32 to 23

##### 7. Data Splitting

```python
# Separate features and target variable
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

##### 8. Feature Scaling

```python
# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

#### 9. Application of Various Machine Learning Models

##### LogisticRegression

```python
# Import necessary libraries for Logistic Regression, hyperparameter tuning, and model evaluation
from sklearn.linear_model import LogisticRegression # importing LogisticRegression model
from sklearn.model_selection import GridSearchCV # for Hyperparameter tuning
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report # for model performance prediction

# Initialize LogisticRegression
log_reg = LogisticRegression()

# Define the parameter grid for GridSearchCV
parameters = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.001, 0.01, 0.1, 1, 10],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'saga'],
    'max_iter': [100, 200, 300]
}

# Perform grid search with cross-validation
grid_search_log_reg = GridSearchCV(log_reg, parameters, cv=5, n_jobs=-1, verbose=1)
grid_search_log_reg.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters for Logistic Regression:", grid_search_log_reg.best_params_)
print("Best Score for Logistic Regression:", grid_search_log_reg.best_score_)

# Use the best estimator from grid search
best_log_reg = grid_search_log_reg.best_estimator_

# Make predictions with the best model
y_train_pred = best_log_reg.predict(X_train)
y_test_pred = best_log_reg.predict(X_test)

# Evaluate the model
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
log_reg_acc=accuracy_score(y_test, y_test_pred)
print("Test Accuracy:",log_reg_acc )
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
```

##### K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.model_selection import GridSearchCV
#from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize KNeighborsClassifier
knn = KNeighborsClassifier()

# Define the parameter grid for GridSearchCV
parameters = {
    'n_neighbors': range(1, 21),  # Number of neighbors to use
    'weights': ['uniform', 'distance'],  # Weight function used in prediction
    'p': [1, 2]  # Power parameter for the Minkowski distance metric
}

# Perform grid search with cross-validation
grid_search_knn = GridSearchCV(knn, parameters, cv=5, n_jobs=-1, verbose=1)
grid_search_knn.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters for K-Nearest Neighbors:", grid_search_knn.best_params_)
print("Best Score for K-Nearest Neighbors:", grid_search_knn.best_score_)

# Use the best estimator from grid search
best_knn = grid_search_knn.best_estimator_

# Make predictions with the best model
y_train_pred = best_knn.predict(X_train)
y_test_pred = best_knn.predict(X_test)

# Evaluate the model
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
knn_acc=accuracy_score(y_test, y_test_pred)
print("Test Accuracy:", knn_acc)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
```

##### Support Vector Classification (SVC)

```python
from sklearn.svm import SVC
#from sklearn.model_selection import GridSearchCV   
# from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize SVC with probability=True to enable probability estimates
svc= SVC(probability=True)

# Define the parameter grid for GridSearchCV
parameters = {
    'gamma': [0.0001, 0.001, 0.01, 0.1],
    'C':[0.01, 0.05, 0.5, 0.1, 1,10, 15,20]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(svc, parameters)
grid_search.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Reinitialize SVC with the best parameters
svc = SVC(C=15, gamma=0.01, probability=True)
svc.fit(X_train, y_train)

# Make predictions
y_pred = svc.predict(X_test)

# Evaluate the model
print("Training Accuracy:", accuracy_score(y_train, svc.predict(X_train)))
svc_acc=accuracy_score(y_test, svc.predict(X_test))
print("Test Accuracy:", svc_acc)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

##### DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
# from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize DecisionTreeClassifier
dtc = DecisionTreeClassifier()

# Define the parameter grid for GridSearchCV
parameters = {
    'criterion':['gini','entropy'],
    'max_depth':range(2,32,1),
    'min_samples_leaf':range(1,10,1),
    'min_samples_split':range(2,10,1),
    'splitter':['best','random']
}

# Perform grid search with cross-validation
grid_search_dt = GridSearchCV(dtc, parameters, cv=5, n_jobs=-1, verbose=1)
grid_search_dt.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters for Decision Tree:", grid_search_dt.best_params_)
print("Best Score for Decision Tree:", grid_search_dt.best_score_)

# Reinitialize DecisionTreeClassifier with the best parameters
best_dtc = DecisionTreeClassifier(**grid_search_dt.best_params_)
best_dtc.fit(X_train, y_train)

# Make predictions
y_pred = best_dtc.predict(X_test)
#y_pred

# Evaluate the model
print("Training Accuracy for Decision Tree:", accuracy_score(y_train, best_dtc.predict(X_train)))
dtc_acc=accuracy_score(y_test, best_dtc.predict(X_test))
print("Test Accuracy for Decision Tree:", dtc_acc)
print("Confusion Matrix for Decision Tree:\n", confusion_matrix(y_test, y_pred))
print("Classification Report for Decision Tree:\n", classification_report(y_test, y_pred))
```

##### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize RandomForestClassifier
rand_clf = RandomForestClassifier()

# Define the parameter grid for GridSearchCV
parameters = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'max_features': ['auto', 'sqrt', 'log2', 0.5],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'n_estimators': [50, 100, 130, 200]
}

# Perform grid search with cross-validation
grid_search_rf = GridSearchCV(rand_clf, parameters, cv=5, n_jobs=-1, verbose=1)
grid_search_rf.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters for Random Forest:", grid_search_rf.best_params_)
print("Best Score for Random Forest:", grid_search_rf.best_score_)

# Use the best estimator from grid search
best_rand_clf = grid_search_rf.best_estimator_

# Make predictions
y_pred = best_rand_clf.predict(X_test)

# Evaluate the model
print("Training Accuracy for Random Forest:", accuracy_score(y_train, best_rand_clf.predict(X_train)))
rand_clf_acc=accuracy_score(y_test, y_pred)
print("Test Accuracy for Random Forest:", rand_clf_acc)
print("Confusion Matrix for Random Forest:\n", confusion_matrix(y_test, y_pred))
print("Classification Report for Random Forest:\n", classification_report(y_test, y_pred))

```

##### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
# from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize GradientBoostingClassifier
gbc = GradientBoostingClassifier()

# Define the parameter grid for GridSearchCV
parameters = {
    'loss': ['deviance', 'exponential'],
    'learning_rate': [0.001, 0.1],
    'n_estimators': [100, 150, 180]
}

# Perform grid search with cross-validation
grid_search_gbc = GridSearchCV(gbc, parameters, cv=2, n_jobs=-1, verbose=1)
grid_search_gbc.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters for Gradient Boosting:", grid_search_gbc.best_params_)

# Reinitialize GradientBoostingClassifier with the best parameters
best_gbc = GradientBoostingClassifier(**grid_search_gbc.best_params_)
best_gbc.fit(X_train, y_train)

# Make predictions
y_pred = best_gbc.predict(X_test)

# Evaluate the model
print("Training Accuracy for Gradient Boosting:", accuracy_score(y_train, best_gbc.predict(X_train)))
gbc_acc=accuracy_score(y_test, y_pred)
print("Test Accuracy for Gradient Boosting:", gbc_acc)
print("Confusion Matrix for Gradient Boosting:\n", confusion_matrix(y_test, y_pred))
print("Classification Report for Gradient Boosting:\n", classification_report(y_test, y_pred))
```

##### XGBClassifier

```python
from xgboost import XGBClassifier
# from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize XGBClassifier
xgb = XGBClassifier(objective='binary:logistic')

# Define the parameter grid for GridSearchCV
parameters = {
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 150, 180],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

# Perform grid search with cross-validation
grid_search_xgb = GridSearchCV(xgb, parameters, cv=3, n_jobs=-1, verbose=1)
grid_search_xgb.fit(X_train, y_train)

# Output the best parameters and best score from grid search
print("Best Parameters for XGBClassifier:", grid_search_xgb.best_params_)
print("Best Score for XGBClassifier:", grid_search_xgb.best_score_)

# Reinitialize XGBClassifier with the best parameters
best_xgb = XGBClassifier(**grid_search_xgb.best_params_)
best_xgb.fit(X_train, y_train)

# Make predictions
y_pred = best_xgb.predict(X_test)

# Evaluate the model
print("Training Accuracy for XGBClassifier:", accuracy_score(y_train, best_xgb.predict(X_train)))
xgb_acc=accuracy_score(y_test, y_pred)
print("Test Accuracy for XGBClassifier:", xgb_acc)
print("Confusion Matrix for XGBClassifier:\n", confusion_matrix(y_test, y_pred))
print("Classification Report for XGBClassifier:\n", classification_report(y_test, y_pred))
```

##### ROC Curves for Various Models 

```python
from sklearn import metrics
plt.figure(figsize=(8,5))
models = [
{
    'label': 'LR',
    'model': log_reg,
},
{
    'label': 'DT',
    'model': dtc,
},
{
    'label': 'SVM',
    'model': svc,
},
{
    'label': 'KNN',
    'model': knn,
},
{
    'label': 'XGBoost',
    'model': xgb,
},
{
    'label': 'RF',
    'model': rand_clf,
},
{
    'label': 'GBDT',
    'model': gbc,
}
]
for m in models:
    model = m['model'] 
    model.fit(X_train, y_train) 
    y_pred=model.predict(X_test) 
    fpr1, tpr1, thresholds = metrics.roc_curve(y_test, model.predict_proba(X_test)[:,1])
    auc = metrics.roc_auc_score(y_test,model.predict(X_test))
    plt.plot(fpr1, tpr1, label='%s - ROC (area = %0.2f)' % (m['label'], auc))

plt.plot([0, 1], [0, 1],'r--')
plt.xlim([-0.01, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1 - Specificity (False Positive Rate)', fontsize=12)
plt.ylabel('Sensitivity (True Positive Rate)', fontsize=12)
plt.title('ROC - Breast Cancer Prediction', fontsize=12)
plt.legend(loc="lower right", fontsize=12)
plt.savefig("roc_breast_cancer.jpeg", format='jpeg', dpi=400, bbox_inches='tight')
plt.show()
```

##### Accuracy and Performance Evaluation of Machine Learning Models

```python
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
models = [
{
    'label': 'LR',
    'model': log_reg,
},
{
    'label': 'DT',
    'model': dtc,
},
{
    'label': 'SVM',
    'model': svc,
},
{
    'label': 'KNN',
    'model': knn,
},
{
    'label': 'XGBoost',
    'model': xgb,
},
{
    'label': 'RF',
    'model': rand_clf,
},
{
    'label': 'GBDT',
    'model': gbc,
}
]

means_roc = []
means_accuracy = [100*round(log_reg_acc,4), 100*round(dtc_acc,4), 100*round(svc_acc,4), 100*round(knn_acc,4), 100*round(xgb_acc,4), 
                  100*round(rand_clf_acc,4), 100*round(gbc_acc,4)]

for m in models:
    model = m['model'] 
    model.fit(X_train, y_train) 
    y_pred=model.predict(X_test) 
    fpr1, tpr1, thresholds = metrics.roc_curve(y_test, model.predict_proba(X_test)[:,1])
    auc = metrics.roc_auc_score(y_test,model.predict(X_test))
    auc = 100*round(auc,4)
    means_roc.append(auc)

print(means_accuracy)
print(means_roc)

# data to plot
n_groups = 7
means_accuracy = tuple(means_accuracy)
means_roc = tuple(means_roc)

# create plot
fig, ax = plt.subplots(figsize=(8,5))
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_accuracy, bar_width,
alpha=opacity,
color='mediumpurple',
label='Accuracy (%)')

rects2 = plt.bar(index + bar_width, means_roc, bar_width,
alpha=opacity,
color='rebeccapurple',
label='ROC (%)')

plt.xlim([-1, 8])
plt.ylim([70, 104])

plt.title('Performance Evaluation - Breast Cancer Prediction', fontsize=12)
plt.xticks(index, ('   LR', '   DT', '   SVM', '   KNN', 'XGBoost' , '   RF', '   GBDT'), rotation=40, ha='center', fontsize=12)
plt.legend(loc="upper right", fontsize=10)
plt.savefig("PE_breast_cancer.jpeg", format='jpeg', dpi=400, bbox_inches='tight')
plt.show()
```
